```python
import requests
import re
import math
import pandas as pd
from lxml import etree
```

```python
# websites url and request headers
url = input('Please enter the link from metacritic: ')
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36 Edg/99.0.1150.39'
}
```

```python
# testing on get request
response = requests.get(url = url, headers = headers)
print(response.text)
```

```python
# retrieve the review count
xml = etree.HTML(text=response.text)
tag = '//a[@class="see_all boxed oswald" and @href="'+url.replace('https://www.metacritic.com', '')+'/user-reviews"]/text()'
print(tag)
```

```python
result = xml.xpath(tag)
print(result)
```

```python
result_str = ''.join(result)
print(result_str)
```

```python
result_int = re.findall('[0-9]+', result_str)
count = math.ceil(int(result_int[0])/100)
print(count)
```

```python
# scrape the reviews
review_list = []
for i in range(count):
    url1 = url+'/user-reviews?page='+str(i)
    response = requests.get(url = url1, headers = headers)
    xml = etree.HTML(text=response.text)
    items = xml.xpath('//div[@class="review pad_top1"]')
    
    for i in items:
        rate = i.xpath('.//div[@class="left fl"]/div/text()')
        name = i.xpath('.//span[@class="author"]/a/text()')
        date = i.xpath('.//span[@class="date"]/text()')
        thumbs_up = i.xpath('.//span[@class="thumb_up"]/span/text()')
        thumbs_down = i.xpath('.//span[@class="thumb_down"]/span/text()')
        if not i.xpath('.//span[@class="blurb blurb_expanded"]/text()'):
            review = i.xpath('.//div[@class="review_body"]/span/text()')
        else:
            review = i.xpath('.//span[@class="blurb blurb_expanded"]/text()')

        dic = {
            'name': name,
            'date': date,
            'review': review,
            'up': thumbs_up,
            'down': thumbs_down
        }
        review_list.append(dic)
```

```python
# examine review length
print(len(review_list))
```

```python
# construct data frame
import pandas as pd
df = pd.DataFrame(review_list)
df
```
